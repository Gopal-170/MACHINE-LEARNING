```python
import numpy as np
import pandas as pd
import sklearn
```

```python
# importing boston dataset

from sklearn.datasets import fetch_openml
df = fetch_openml(name="california_housing", as_frame=True)
```

```python
print(df)
```

```python
# will convert data into pandas dataframe
df = pd.DataFrame(df.data, columns=df.feature_names)
df.head()
```

```python
# exporting dataframe in csv file

df.to_csv('california_housing.csv', index=False)
```

```python
#  information about dataframe

df.info()
```

```python
# no missing vaues
df.isnull().sum()
```

```python
# group the values based on mean

df.groupby(['households']).mean(numeric_only=True)
```
